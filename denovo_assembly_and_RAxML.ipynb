## 3. Assemble the Data
The data need to be put together and placed on the reference genome for *Q. robur*.

```python
import ipyrad as ip
data = ip.Assembly('Quercus') # create an Assembly object
```

```python
# 'data.set_params('project_dir', 'assembly')' giving errors, saying there is a space in the name...

data.set_params('sorted_fastq_path', '~/scratch-dsi/rawdata/*.gz')
data.set_params('mindepth_statistical', '5')
data.set_params('mindepth_majrule', '2')
data.set_params('project_dir', '~/scratch-dsi/DataAnalysisandNotebooks/quercus/')
data.set_params('output_formats', 'lpsvaku')
# data.set_params('output_formats', 'nlpsvaku') or *?
data.get_params()
```

```python
data.run('12', force=True)
```

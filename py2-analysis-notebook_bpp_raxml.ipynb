# BPP Analysis

About BPP: a Bayesian Markov chain Monte Carlo analysis tool for inferring the best species tree from sequence data. Priors are estimated and adjusted until the average optimal set of parameters are found for the dataset. The data is analyzed under the multi-species coalescent when species delimitation are fixed. The analysis can be done either using a fixed or unfixed guide tree. Because the phylogeny of the live oaks is relatively well resolved, a fixed newick tree guide was used. This also reduces the amount of time for the analysis to run. 

Two two parameters estimated in this analysis are tau and theta. Tau corresponds to species divergence times and theta with population size paramters of ancestral and modern populations/species

# Drawing plots for the MCMC parameters of the no-data and with-data analyses. 
Are these results divergent enough to say that the sequence data is informative?

```python
## get full MCMC table from the .files attribute
table_1 = pd.read_csv(ndata.files.mcmcfiles[0], sep='\t', index_col=0)
table_2 = pd.read_csv(bdata.files.mcmcfiles[0], sep='\t', index_col=0)

## draw barplots
canvas = toyplot.Canvas(width=800, height=800)

for aidx in range(table_1.shape[1]):
    ## get data
    param = table_1.columns[aidx]
    hist1 = np.histogram(table_1[param], density=True)
    hist2 = np.histogram(table_2[param], density=True)

    ## build plot
    nx = int(np.ceil(np.sqrt(table_1.shape[1])))
    axes = canvas.cartesian(grid=(nx, nx, aidx), margin=25)
    axes.bars(hist1, opacity=0.65, style={"stroke-width": 0})
    axes.bars(hist2, opacity=0.65, style={"stroke-width": 0})

    ## style axes (hide tick labels for clarity)
    axes.label.text = param[:15]
    axes.label.style["font-size"] = "12px"
    axes.x.ticks.labels.show = False
    axes.y.ticks.labels.show = False
    
```

```python
# save the plots in the folder that contains the analysis notebooks and data
import toyplot.pdf
toyplot.pdf.render(canvas, "/Users/maiahernandez/Desktop/SeniorThesis/DataAnalysisandNotebooks/quercus-analysis/analysis-bpp/bpp.pdf")
```

## 2. Download the Sequence Data

Sequence data are from Eaton et al. 2015. The dataset includes *Quercus* ser *Virentes* sequence reads, as well as outgroup samples. The accession number used for downloading the sra files: 'SRP055977'.

```python
import ipyrad.analysis as ipa
import pandas as pd

sra = ipa.sratools(accession="SRP055977", workdir="rawdata")
```

```python
# from: http://eaton-lab.org/posts/sra-downloads/
# 'fetching' the SRA project assemblies
proj = sra.fetch_runinfo()
```

```python
proj
```

```python
proj[["Run", "ScientificName"]]
```

```python
# download the raw data files. Files were downloaded into a file named 'rawdata'
sra.run(name_fields=(30,1), name_separator='_')
```

## 3. Assemble the Data
The data need to be put together and placed on the reference genome for *Q. robur*.

```python
import ipyrad as ip

data = ip.Assembly('Quercus') # create an Assembly object
```

```python
# 'data.set_params('project_dir', 'assembly')' giving errors, saying there is a space in the name...
data.set_params('sorted_fastq_path', 'rawdata/*.gz')
data.set_params('mindepth_statistical', '5')
data.set_params('mindepth_majrule', '2')

data.get_params()
```

```python
data.run('12')
```
